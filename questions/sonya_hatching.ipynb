```python
import iris
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
infile = '~/Downloads/data/evspsbl_Amon_CSIRO-Mk3-6-0_historicalMisc_r1i1p4_185001-201212.nc'
```

```python
cube = iris.load_cube(infile)
```

```python
print cube
```

```python
pvals = cube[0, ::].data * 1000
```

```python
x = cube.coord('longitude').points
y = cube.coord('latitude').points
```

```python
clevs = [0.05,0.075]
plt.contourf(x, y, pvals)
plt.colorbar()

plt.contourf(x, y, pvals, clevs, colors='none', hatches=['///'], extend='upper', zord=3)
plt.show()
```

```python
clevs = [0.05,0.075]
plt.contour(x, y, pvals, colors='grey', linestyles='-', zord=2)
plt.contourf(x, y, pvals, clevs, colors='none', hatches=['///'], extend='upper', zord=3)
plt.show()
```
